```python
import numpy as np
from scipy import misc
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import trange, tqdm
from numpy.fft import fft2 as FFT, ifft2 as iFFT
from numpy.fft import rfft2 as rFFT, irfft2 as irFFT
from numpy.fft import fftn as FFTN, ifftn as iFFTN
from numba import jit
#from multiprocessing import Pool as ThreadPool
from pathos.multiprocessing import ProcessPool
from psutil import cpu_count
from operator import sub
```

```python
#参考
#https://algorithm.joho.info/programming/python/numpy-scipy-fft-spectrum/
img = misc.imread('../../img/hari.jpeg', mode='L')
plt.imshow(img, 'gray')
```

```python
# 高速フーリエ変換(2次元)
fimg = np.fft.fft2(img)
plt.plot(fimg)
plt.show()
```

```python
# 零周波数成分を配列の左上から中心に移動
fimg =  np.fft.fftshift(fimg)
plt.plot(fimg)
plt.show()
```

```python
# 強度スペクトルの計算
mag = 20*np.log(np.abs(fimg))
plt.plot(mag)
plt.show()
```
