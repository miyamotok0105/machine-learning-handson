# 数値微分

微分はある瞬間の変化量

$ \frac{df(x)}{dx}=\lim_{h \to 0} \frac{f(x+h)-f(x)}{h} $

```python
def numerical_diff2(f, x):
    h = 1e-4
    return (f(x+h) - f(x)) / h

def numerical_diff3(f, x):
    h = 1e-4
    return (f(x+h) - f(x-h)) / (2*h)

def sample_func(x):
    return x**2

x = np.arange(-7.0, 7.0, 0.1)
x2 = np.arange(-1.0, 7.0, 0.1)
x3 = np.arange(-1.0, 7.0, 0.1)
k = 2

y = sample_func(x)
d2 = numerical_diff2(sample_func, k)
d3 = numerical_diff3(sample_func, k)
y2 = d2 * x2 + sample_func(k) - d2*k
y3 = d3 * x3 + sample_func(k) - d3*k

plt.plot(x, y) #曲線
plt.plot(x2, y2) 
plt.plot(x3, y3) 
plt.show()
```
