# 数値微分の線をプロットする

```python
#数値微分
def numerical_diff(f, x):
    h = 1e-4 # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)
#2次関数
def function_1(x):
    return 0.01*x**2 + 0.1*x 
#数値微分で傾きを、f(x) - dxで切片を求める
def tangent_line(f, x):
    d = numerical_diff(f, x)
    print(d)
    #y = dx + bのbの切片を求めたい。
    #b = y - dx
    #b = f(x) - dx
    #なのでf(x) - dxで切片を求める
    #f(x)が2次関数。dは数値微分した値。
    y = f(x) - d*x
    #tはラムダ関数の引数。returnで関数を返す。
    #傾き d 切片 y の 1次関数を返す関数
    return lambda t: d*t + y
     
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")

tf = tangent_line(function_1, 5)
y2 = tf(x)

plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

# 1次関数の傾きをプロット

# 中央の傾きを使用する

数値微分で(f(x) - f(x-h))を使うと傾きがずれてるので、(f(x+h) - f(x-h))を使用して中央にズラす。

```python
def numerical_diff_before(f, x):
    h = 1e-4 # 0.0001
    return (f(x) - f(x-h)) / (h)

def numerical_diff_center(f, x):
    h = 1e-4 # 0.0001
    return ((f(x+h) - f(x-h)) / (2*h))

def numerical_diff_after(f, x):
    h = 1e-4 # 0.0001
    return ((f(x+h) - f(x)) / (h))

def function_1(x):
    return 0.01*x**2 + 0.1*x 

def tangent_before_line(f, x):
    d = numerical_diff_before(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y

def tangent_center_line(f, x):
    d = numerical_diff_center(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y

def tangent_after_line(f, x):
    d = numerical_diff_after(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y

x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")

tf_before = tangent_before_line(function_1, 5)
y2_before = tf_before(x)
tf_center = tangent_center_line(function_1, 5)
y2_center = tf_center(x)
tf_after = tangent_after_line(function_1, 5)
y2_after = tf_after(x)

print("y2", y2[:4])
plt.plot(x, y)
plt.plot(x, y2_before)
plt.plot(x, y2_center)
plt.plot(x, y2_after)
plt.show()
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
from plotly.graph_objs import *
import numpy as np
init_notebook_mode(connected=True)

x, y, z = np.random.multivariate_normal(np.array([0,0,0]), np.eye(3), 200).transpose()
trace1 = go.Scatter3d(x=x, y=y, z=z, mode='markers',
    marker=dict(size=12, line=dict(color='rgba(217, 217, 217, 0.14)', width=0.5), opacity=0.8))

x2, y2, z2 = np.random.multivariate_normal(np.array([0,0,0]), np.eye(3), 200).transpose()
trace2 = go.Scatter3d(x=x2, y=y2, z=z2, mode='markers', marker=dict(color='rgb(127, 127, 127)', size=12,  symbol='circle', line=dict(color='rgb(204, 204, 204)', width=1),opacity=0.9))

data = [trace1, trace2]
layout = go.Layout(margin=dict(l=0, r=0, b=0, t=0))
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='simple-3d-scatter')
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
from plotly.graph_objs import *
import numpy as np
init_notebook_mode(connected=True)

def function_2(x, a1, a2):
    return a1*x[0]**2 + a2*x[1]**2

x = np.meshgrid(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1))
data = [
    go.Surface(x=x[0], y=x[1], z=function_2(x, 2, 2), colorscale='Viridis'),
    go.Surface(x=x[0], y=x[1], z=function_2(x, 2, 1), colorscale='red'),
    go.Surface(x=x[0], y=x[1], z=function_2(x, 2, 0))
]
layout = go.Layout(title='plotly', autosize=False, width=500, height=500, margin=dict(l=65, r=50, b=65, t=90))
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='3d-surface')
```
