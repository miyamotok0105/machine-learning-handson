```python
# -*- coding: utf-8 -*-
import matplotlib
import matplotlib.pyplot as plt
import cv2
import os
import glob
import random
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np
from skimage import data
from skimage.filters import try_all_threshold
from scipy import misc
img = cv2.imread('../img/hari.jpeg')
filename = "hari.jpeg"
filename = filename.replace(".jpg", "")
img = ia.imresize_single_image(img, (256, 256))
print(img.shape) #640, 856, 3
sometimes = lambda aug: iaa.Sometimes(0.9, aug)

for i in range(2):
    scale_x1 = random.uniform(0.9,1.2)
    scale_x2 = random.uniform(0.9,1.2)
    scale_y1 = random.uniform(0.9,1.2)
    scale_y2 = random.uniform(0.9,1.2)
    aug = sometimes(iaa.Affine(scale={"x": (scale_x1, scale_x2), "y": (scale_y1, scale_y2)}, cval=(0, 255)))
    plt.imshow(cv2.cvtColor(aug.augment_image(img), cv2.COLOR_BGR2RGB),cmap='gray')
    plt.show()
```

```python
for i in range(2):
    scale1 = random.uniform(0,0.02)
    aug = sometimes(iaa.PiecewiseAffine(scale=(0.00, scale1), cval=(0, 255)))
    plt.imshow(aug.augment_image(img),cmap='gray')
    plt.show()

```

```python

for i in range(2):
    rotate1 = random.uniform(-5,5)
    aug = sometimes(iaa.Affine(rotate=(-20, rotate1), cval=(0, 255)))
    plt.imshow(aug.augment_image(img),cmap='gray')
    plt.show()

```

```python

for i in range(2):
    shear1 = random.uniform(-5,5)
    aug = sometimes(iaa.Affine(shear=(shear1, 5), cval=(0, 255)))
    plt.imshow(aug.augment_image(img),cmap='gray')
    plt.show()
```

# グレースケール

```python
# -*- coding: utf-8 -*-
import matplotlib
import matplotlib.pyplot as plt
import cv2
import os
import glob
import random
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np
from skimage import data
from skimage.filters import try_all_threshold
from scipy import misc
img = cv2.imread('../img/hari.jpeg')
filename = "hari.jpeg"
filename = filename.replace(".jpg", "")
img = ia.imresize_single_image(img, (256, 256))
print(img.shape) #640, 856, 3
sometimes = lambda aug: iaa.Sometimes(0.9, aug)

for i in range(2):
    scale_x1 = random.uniform(0.9,1.2)
    scale_x2 = random.uniform(0.9,1.2)
    scale_y1 = random.uniform(0.9,1.2)
    scale_y2 = random.uniform(0.9,1.2)
    aug = sometimes(iaa.Affine(scale={"x": (scale_x1, scale_x2), "y": (scale_y1, scale_y2)}, cval=(0, 255)))
    plt.imshow(cv2.cvtColor(aug.augment_image(img), cv2.COLOR_RGB2GRAY),cmap='gray')
    plt.show()
```

```python

aug = iaa.Invert(1, per_channel=0.1)
plt.imshow(cv2.cvtColor(aug.augment_image(img), cv2.COLOR_RGB2GRAY),cmap='gray')
plt.show()

```

```python

aug  = iaa.Add(10)
plt.imshow(cv2.cvtColor(aug.augment_image(img), cv2.COLOR_RGB2GRAY),cmap='gray')
plt.show()
```

```python

aug = iaa.Sharpen((0.0, 1.0))
plt.imshow(cv2.cvtColor(aug.augment_image(img), cv2.COLOR_RGB2GRAY),cmap='gray')
plt.show()

```
